```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../dataset/Titanic-Dataset.csv")
```

```python
##### SKEWNESS - measures the asymmetry of the distribution
### Skewness = 0 for perfectly symmetrical distribution - mean = median = mode
### Skewness < 0 => skewed to left side (IF THE TAIL IS LONGER ON THE LEFT SIDE) - mean < median < mode
### Skewness > 0 => skewed to right side (IF THE TAIL IS LONGER ON THE RIGHT SIDE) - mode < median < mean
mean = np.mean(dataset['Age'])
dataset.fillna({'Age': np.mean(dataset['Age'])}, inplace=True)
skewness = np.sum([(x-mean)**3 for x in dataset['Age']])/((len(dataset['Age']) - 1) * np.std(dataset['Age'])**3)
```

```python
skewness
```

```python
sns.histplot(x='Age', data=dataset)
plt.show()
```
