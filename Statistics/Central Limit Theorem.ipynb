```python
##### Central Limit Theorem (CLT)
### You may have data which is Normal distributed or non-Normal distributed
### IF you pick up random samples - say S1, S2, S3, S4...; then their means x1, x2, x3, x4, x5... will follow a normal distribution; 
### & mean of x1, x2, x3, x4, x5... will be equal to population mean
# Used when we have a very large population
```

```python
# 1. A random data must be sampled and selected randomly fro the population
# 2. There should not be multicollinearity in the sampled data i.e. one sample should not influence other samples
# 3. Sample size should not be more that 10% of population
# 4. Generally, sample size > 30 is considered good
```

```python
pop_data = [np.random.randint(10,100) for i in range(10000)]
pop_table = pd.DataFrame({"Pop_data": pop_data})
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x="Pop_data", data=pop_table)
plt.show()
```

```python
sam_mean = []
for samples in range(60):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sam_mean.append(np.mean(sample_data))
```

```python
sample_means = pd.DataFrame({"means": sam_mean})
```

```python
sns.kdeplot(x="means", data=sample_means)
plt.show()
```
