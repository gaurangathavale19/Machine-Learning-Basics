```python
### Regularization
# This is a form of regression, that constraints/regularizes or shrinks the coefficient estimates towards zero.
# This technique discourages learning a more complex or flexible model, so as to avoid the risk of overfitting.
```

```python
### Types: 
    # a. Lasso Regularization (L1)
    # b. Ridge Regularization (L2)
```

```python
## a. Lasso Regularization (L1):
# Used in feature selection using a shrinkage method also referred to as penalized regression method.
# Lasso Regularization magnitude of coefficients can be **exactly** zero.
# Cost Function = Loss + lambda*sum(||w||)
# Note: Loss = sum of squared residual
      # lambda = penalty
      # w = slope of the curve = sum (|m1| + |m2| + ... + |mn|)
```

```python
## b. Ridge Regularization (L2)
# Extension to linear regression that introduces a regularization term to reduce model complexity and help prevent overfitting.
# Ridge Regularization is working value/magnitude of coefficients is **almost equal** to zero.
# Cost Function = Loss + lambda*sum(||w||^2)
# Note: Loss = sum of squared residual
      # lambda = penalty
      # w = slope of the curve = sum (|m1| + |m2| + ... + |mn|)

# PROs: Reduces the computational complexity
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
dataset = pd.read_csv(r"../dataset/House Price India.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
x.head(3)
```

```python
y = dataset["Price"]
y.head(3)
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns=x.columns)
x.head(3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Linear Regression

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
print(mean_squared_error(y_test, lr.predict(x_test)))
print(mean_absolute_error(y_test, lr.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, lr.predict(x_test))))
```

```python
plt.figure(figsize=(20,7))
plt.bar(x.columns, lr.coef_)
plt.title("Linear Regression")
plt.xlabel("columns")
plt.ylabel("coef")
plt.show()
```

# Lasso - Feature Selection

```python
l1 = Lasso(alpha=0.5)
l1.fit(x_train, y_train)
l1.score(x_test, y_test)
```

```python
print(mean_squared_error(y_test, l1.predict(x_test)))
print(mean_absolute_error(y_test, l1.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, l1.predict(x_test))))
```

```python
plt.figure(figsize=(20,7))
plt.bar(x.columns, l1.coef_)
plt.title("Lasso")
plt.xlabel("columns")
plt.ylabel("coef")
plt.show()
```

 # Ridge

```python
l2 = Ridge(alpha=10)
l2.fit(x_train, y_train)
l2.score(x_test, y_test)
```

```python
print(mean_squared_error(y_test, l2.predict(x_test)))
print(mean_absolute_error(y_test, l2.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test, l2.predict(x_test))))
```

```python
plt.figure(figsize=(20,7))
plt.bar(x.columns, l2.coef_)
plt.title("Ridge")
plt.xlabel("columns")
plt.ylabel("coef")
plt.show()
```

```python
df = pd.DataFrame({"col_name": x.columns, "LinearRegression": lr.coef_, "Lasso": l1.coef_, "Ridge": l2.coef_})
```
