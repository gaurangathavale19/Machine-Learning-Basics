```python
##### Feature Scaling
### Standardization - It is a very effective technique which re-scales a feature value so that it has distribution with 0 mean value and variance = 1
# Xnew = (Xi - Xmean) / (std)

# Note: Feature scaling doesnt eliminate outliers
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"../dataset/loan_data_set.csv")
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]), columns=["x"])
dataset
```

```python
plt.subplot(1,2,1)
plt.title("Original")
sns.distplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("Scaled")
sns.distplot(dataset["ApplicantIncome_ss"])

plt.show()
```
