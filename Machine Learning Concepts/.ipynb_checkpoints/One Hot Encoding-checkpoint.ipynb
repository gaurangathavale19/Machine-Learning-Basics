```python
##### One Hot Encoding
### Used when the number of categories in a categorical columns is less - eg: Gender, Married
import pandas as pd
```

```python
dataset = pd.read_csv(r"../dataset/loan_data_set.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna({"Gender": dataset["Gender"].mode()[0],
                "Married": dataset["Married"].mode()[0]}, inplace=True)
```

```python
### One Hot Encoding using get_dummies()
encoding_data = dataset[["Gender", "Married"]]
pd.get_dummies(encoding_data) 
### But still data is in boolean format - but we need numbers - Hence, use scikit -> OneHotEncoder
```

```python
### Using sklearn -> OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(encoding_data).toarray()
```

```python
pd.DataFrame(ar, columns=["Gender_Female", "Gender_Male", "Married_No", "Married_Yes"])
```

```python
### To decrease the size of the dataframe - as it has become huge - we can use drop="first"
ohe_drop_first = OneHotEncoder(drop="first")
ar1 = ohe_drop_first.fit_transform(encoding_data).toarray()
ar1
```

```python
pd.DataFrame(ar1, columns=["Gender_Male", "Married_Yes"])
```
