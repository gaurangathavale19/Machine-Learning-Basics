```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"../dataset/loan_data_set.csv")
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
### Outlier Remove
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
iqr = q3-q1
```

```python
min_r = q1-1.5*iqr
max_r = q3+1.5*iqr
min_r, max_r
```

```python
dataset = dataset[(dataset["CoapplicantIncome"] <= max_r) & (dataset["CoapplicantIncome"] >= min_r)]
```

```python
### After outlier removal
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
### Function Transformer
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p) # Can use any function
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
### After transformation
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(dataset["CoapplicantIncome"])
plt.title("Before Transformation")

plt.subplot(1,2,2)
sns.distplot(dataset["CoapplicantIncome_tf"])
plt.title("After Transformation")
plt.show()
```
