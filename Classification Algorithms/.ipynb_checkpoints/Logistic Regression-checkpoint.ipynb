# Logistic Regression

### Types of Logistic Regression
#### 1. Binomial (2 unordered categories) 
#### 2. Multinomial (3 or more unordered categories)
#### 3. Ordinal (3 or more ordered categories)

# Logistic Regression uses sigmoid function
## Sigmoid y = 1/1+e^-x
### y = dependent variable
### x = independent variable
### e = Euler's Constant = 2.71828

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
```

### Logistic Regression - (Binary Classification - single input)

```python
dataset = pd.read_csv(r"../dataset/Social_Network_Ads.csv")
dataset.drop(columns=["User ID", "Gender", "EstimatedSalary"], inplace=True) # Just because we need to try LR on single input
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=dataset)
plt.show()
```

```python
x = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
lr.predict(x_test)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age", y="Purchased", data=dataset)
sns.lineplot(x="Age", y=lr.predict(x), data=dataset, color="red")
plt.show()
```

## Logistic Regression - (Binary Classification - multiple input)

```python
dataset1 = pd.read_csv(r"../dataset/placements.csv")
dataset1 = dataset1.iloc[:, 1:]
```

```python
dataset1.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa", y="iq", data=dataset1, hue="placement")
plt.legend(loc=1)
plt.show()
```

```python
x1 = dataset1.iloc[:, :-1]
y1 = dataset1["placement"]
```

```python
x1_train, x1_test, y1_train, y1_test = tts(x1, y1, test_size=0.2, random_state=42)
```

```python
lr1 = LogisticRegression()
lr1.fit(x1_train, y1_train)
```

```python
lr1.score(x1_test, y1_test)
```

```python
plot_decision_regions(x1.to_numpy(), y1.to_numpy(), clf=lr1)
plt.show()
```
